Question 1 : What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?
-  When we do hypothesis testing, we make decisions based on sample data — but sometimes we can make wrong decisions.
These wrong decisions are called Type I and Type II errors.

. Type I Error (α Error):

Meaning: Rejecting the null hypothesis (H₀) even though it is actually true.

In simple words: You think there is an effect or difference — but actually, there isn’t.

Example:
Suppose a doctor tests a healthy person for a disease and the test says “Positive.”
→ This is a Type I Error (false alarm).

Symbol: α (alpha)

Probability of making Type I error = Significance level (α)

.  Type II Error (β Error):

Meaning: Failing to reject the null hypothesis (H₀) even though it is actually false.

In simple words: You think there’s no effect — but actually, there is.

Example:
The doctor tests a sick person, but the test says “Negative.”
→ This is a Type II Error (missed detection).

Symbol: β (beta)

1 - β = Power of the test (the ability to detect a true effect)

Type I and Type II errors affect decision-making by causing false conclusions. A Type I error leads to acting on a false effect (false positive), while a Type II error causes missing a real effect (false negative). Both can result in wrong business or research decisions, so balancing them is crucial for accurate conclusions.







































Question 2:What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?
-  The P-value (probability value) is the probability of getting a test result as extreme as, or more extreme than, the one observed — assuming the null hypothesis (H₀) is true.

In simple words:
It tells us how strong the evidence is against the null hypothesis.

-  Interpretation:

If the P-value ≤ significance level (α) (for example, α = 0.05),
→ We reject the null hypothesis (H₀).
→ There is strong evidence that the alternative hypothesis (H₁) is true.

If the P-value > α,
→ We fail to reject H₀.
→ There is not enough evidence to support H₁.

. In context of the null hypothesis:

The smaller the P-value, the stronger the evidence against the null hypothesis.
A large P-value means the observed data is likely under H₀, so we keep it.


Question 3:Explain the difference between a Z-test and a T-test, including when to use each ?
-  Both Z-test and T-test are hypothesis tests used to compare sample data with population data or between two samples.
The main difference lies in the sample size and whether the population standard deviation (σ) is known or not.
- When to Use:

Use Z-test:
When you have a large sample and know the population standard deviation.
Example: Comparing factory output to a known average.

Use T-test:
When you have a small sample or don’t know the population standard deviation.
Example: Comparing mean marks of two small student groups.


### Question 4:What is a confidence interval, and how does the margin of error influence its width and interpretation?
-  A confidence interval (CI) is a range of values estimated from sample data that is likely to contain the true population parameter (like mean or proportion) with a certain level of confidence.

Common confidence levels: 90%, 95%, 99%

Example: A 95% CI for a mean of 50 might be [48, 52], meaning we are 95% confident the true mean lies between 48 and 52.

- Margin of Error (MoE):

The margin of error measures the uncertainty or variability in the estimate.

Formula:

CI=Sample estimate ± Margin of Error

Influence on width:

Larger margin of error → wider interval → more uncertainty

Smaller margin of error → narrower interval → more precise estimate

Interpretation:
The CI tells us the range in which the true population value is likely to fall. The margin of error shows how precise that estimate is.


Question 5: Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?

-Purpose of ANOVA:

ANOVA (Analysis of Variance) is used to compare the means of three or more groups to see if at least one group mean is significantly different from the others.

Instead of doing multiple t-tests (which increases error), ANOVA tests all groups simultaneously.

Null hypothesis (H₀): All group means are equal.

Alternative hypothesis (H₁): At least one group mean is different.

- Assumptions of ANOVA:

Independence: Observations are independent of each other.

Normality: Data in each group is approximately normally distributed.

Homogeneity of variance: The variance across groups is roughly equal.

- How it extends hypothesis testing:

Standard t-tests compare two groups at a time.

ANOVA allows testing three or more groups together in one test, avoiding multiple comparisons and reducing the risk of Type I error.

If ANOVA shows a significant result, post-hoc tests (like Tukey’s HSD) are used to find which specific groups differ.

Question 6: Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.
-  A one-sample Z-test is used to check whether the mean of a sample differs significantly from a known population mean, assuming the population standard deviation is known.


```python
from statsmodels.stats.weightstats import ztest

import numpy as np

# Example dataset: Sample data
data = np.array([102, 100, 98, 105, 110, 95, 99, 101, 104, 97])

# Known population mean (for example, mu = 100)
population_mean = 100

# Perform one-sample Z-test
z_statistic, p_value = ztest(data, value=population_mean)

# Print results
print("Z-Statistic:", z_statistic)
print("P-Value:", p_value)

# Interpret the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
else:
    print("Fail to reject the null hypothesis: Sample mean is NOT significantly different from population mean.")
```

Question 8: Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.
-  The CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population’s original distribution.

We’ll use a non-normal distribution (e.g., exponential distribution) for this example.


```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a non-normal population (exponential distribution)
population_size = 10000
population = np.random.exponential(scale=2, size=population_size)  # scale = 1/lambda

# Step 2: Parameters for sampling
sample_size = 50   # Number of observations per sample
num_samples = 1000 # Number of samples

# Step 3: Generate sample means
sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(num_samples)]

# Step 4: Plot the original population distribution
plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
plt.hist(population, bins=50, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Original Population Distribution (Exponential)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Step 5: Plot the sampling distribution of the mean
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title(f"Sampling Distribution of Mean (n={sample_size})")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

# Print mean comparison
print("Population mean:", np.mean(population))
print("Mean of sample means:", np.mean(sample_means))
```

Question 9: Write a Python function to calculate and visualize the confidence interval for a sample mean.



```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def confidence_interval(data, confidence=0.95):
    """
    Calculate and visualize the confidence interval for the sample mean.

    Parameters:
    data : array-like
        Sample data
    confidence : float
        Confidence level (default 0.95)

    Returns:
    mean : float
        Sample mean
    lower_bound : float
        Lower bound of CI
    upper_bound : float
        Upper bound of CI
    """
    # Convert to numpy array
    data = np.array(data)

    # Sample statistics
    mean = np.mean(data)
    n = len(data)
    std_err = stats.sem(data)  # Standard error of the mean

    # Margin of error
    h = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)

    # Confidence interval
    lower_bound = mean - h
    upper_bound = mean + h

    # Print results
    print(f"Sample mean: {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

    # Visualization
    plt.figure(figsize=(6,4))
    plt.errorbar(1, mean, yerr=h, fmt='o', color='blue', capsize=10)
    plt.xlim(0, 2)
    plt.xticks([])
    plt.ylabel("Value")
    plt.title(f"{int(confidence*100)}% Confidence Interval for Sample Mean")
    plt.show()

    return mean, lower_bound, upper_bound

# Example usage
sample_data = [102, 100, 98, 105, 110, 95, 99, 101, 104, 97]
confidence_interval(sample_data, confidence=0.95)
```

Question 10: Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.



```python
import numpy as np
from scipy.stats import chisquare

# Step 1: Observed frequencies (example data)
observed = np.array([50, 30, 20])  # e.g., counts of outcomes in categories A, B, C

# Step 2: Expected frequencies (based on theoretical distribution)
expected = np.array([40, 40, 20])  # e.g., expected counts if null hypothesis is true

# Step 3: Perform Chi-square goodness-of-fit test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Step 4: Print results
print("Chi-square Statistic:", chi_stat)
print("P-Value:", p_value)

# Step 5: Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Observed distribution significantly differs from expected.")
else:
    print("Fail to reject the null hypothesis: Observed distribution does NOT significantly differ from expected.")
```
